```python
input_file = 'spoon_river_anthology.txt'
output_file = 'spoon_river.json'

import json

class SpoonInfo:    
    def __init__(self, name, poem, position):
        self.name = name
        self.poem = poem
        self.position = position
        self.references = []
        self.referenced_by = []

infile =  open(input_file, 'r')

line_counter = 0
def readLine(label = ''):
    global  line_counter, fd
    l = infile.readline().strip()
    line_counter += 1
    #print "%04i - %s %s" % (line_counter, l, label)
    return l


data = []
index = 0;
readLine('[blank - 0]')
while True:
    readLine('[blank - 1]')
    readLine('[blank - 2]')
    
    name = readLine('[name]')
    if not name : break
        
    readLine('[blank]')
    
    poem = []
    line = readLine('[poem - %i] ' % len(poem))
    while line:
        poem.append(line)
        line = readLine('[poem - %i] ' % len(poem))
        
    info = SpoonInfo(name, poem, index)
    data.append(info)
    index += 1
    
print "Closing %s" % input_file
infile.close()

for d1 in data:
    name1 = d1.name
    poem1 = " ".join(d1.poem)
    for d2 in data:
        if d1 is d2:
            continue
        name2 = d2.name
        if name2 in poem1:
            print "%s -> %s" % (name1, name2)
            d1.references.append(name2)
            d2.referenced_by.append(name1)

print "Saving %s" % output_file
with open(output_file, 'w') as outfile:
    json.dump([ob.__dict__ for ob in data], outfile)

    
```
